```python
import pandas as pd
from pandas import DataFrame
import seaborn as sns
from matplotlib import pyplot as plt
```

Read and convert data

```python
path = "./Resources/SOCR-HeightWeight.csv"
raw_data = pd.read_csv(path)
raw_data.head(10)
raw_data["Height(Centimeter)"] = raw_data["Height(Inches)"] * 2.54
raw_data["Weight(Kilogram)"] = raw_data["Weight(Pounds)"] * 0.45359237

data = raw_data.filter(["Height(Centimeter)","Weight(Kilogram)"])
```

Visualize Data

```python
fig, axs = plt.subplots(1,3,figsize=(20,5))
sns.scatterplot(data=data,x=data["Weight(Kilogram)"],y=data["Height(Centimeter)"],ax=axs[0])
sns.histplot(data["Height(Centimeter)"], ax = axs[1])
sns.histplot(data["Weight(Kilogram)"],ax=axs[2])
```
